```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
cars_df=pd.read_csv('/content/mtcars.csv')
print(cars_df)

a=plt.scatter(cars_df["mpg"],cars_df["wt"],marker="*")
print(a)
plt.xlabel("MPG")
plt.ylabel("Weight")
plt.title("relationsip  between weight of cars and mpg")
plt.show()

trans=cars_df["am"].value_counts()
print(trans)
trans.plot.bar()
plt.xlabel("transmission")
plt.title("frequency distribution of cars")
plt.show()

cars_df["mpg"].plot(kind='box')
plt.xlabel("mpg")
plt.title("box and wisker plot to mpg")
plt.show()

min_value=np.min(cars_df['mpg'])
q1=np.percentile(cars_df['mpg'],25)
median=np.median(cars_df['mpg'])
q3=np.percentile(cars_df['mpg'],75)
max_value=np.max(cars_df['mpg'])
print("minvalue:",min_value)
print("q1:",q1)
print("median:",median)
print("q3:",q3)
print("maxvalue:",max_value)
```
