```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
cars_df=pd.read_csv("/content/Cars93 (1).csv")
print(cars_df)

fig, ax = plt.subplots(2, 2, figsize=(10, 7)) # Initialize subplots

fig.suptitle("Many Boxplots", fontsize=16) # Use suptitle for the main title

ax[0][0].boxplot(cars_df["Price"][cars_df["Type"]=="Compact"]) # Correct column name and capitalization
ax[0][0].set_title("Compact") # Use set_title for subplot titles

ax[0][1].boxplot(cars_df["Price"][cars_df["Type"]=="Large"]) # Correct column name and capitalization
ax[0][1].set_title("Large") # Use set_title for subplot titles

ax[1][0].boxplot(cars_df["Price"][cars_df["Type"]=="Van"]) # Correct column name and capitalization
ax[1][0].set_title("Van") # Use set_title for subplot titles

ax[1][1].boxplot(cars_df["Price"][cars_df["Type"]=="Midsize"]) # Correct column name and capitalization
ax[1][1].set_title("Midsize") # Use set_title for subplot titles


plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust rect to leave space for the main title
plt.show()

min_value=np.min(cars_df["Price"]) # Correct column name and capitalization
ni=np.percentile(cars_df["Price"],25) # Correct column name and capitalization
median=np.median(cars_df['Price']) # Correct column name and capitalization
de=np.percentile(cars_df["Price"],75) # Correct column name and capitalization
max_value=np.max(cars_df["Price"]) # Correct column name and capitalization
print("mnimum value",min_value)
print("ni",ni)
print("median",median)
print("de",de)
print("maximum value",max_value)
```
